### Gradient Descent and backtracking line search

```python
def fx(x):
    return  x**4 - (1.3*(x**3))-(1.95*(x**2))+ 4*x+3.65
def diff_fx(x):
    return (4*(x**3))-(3.9*(x**2))-(3.9*x)+4
```

using a step size (mu) of 0.1 (mu E 0,1) xo = 2. We find the number of iterations by making use of a while loop and count variable to store each iteration

```python
def gradient_descent(fx,xo,alpha,eps):
    temp = xo - alpha * diff_fx(xo)
    count = 0
    while fx(xo) - fx(temp) > eps:
        xo = temp
        temp = xo - alpha * diff_fx(xo)
        count = count+1
    print("The minimum value of fx is:",round(xo)) #ensure the minimum is equal to -1 (assert test can also be used)
    print("The minimum number of iterations required to descend to the minimum at alpha = ",alpha," is:",count)

    
```

```python
gradient_descent(fx,2,0.1,0.000001)
```

Try again with the same step length, but from the startingpointx0= 1.5. Why does it take more iterations to achieve the targetaccuracy, although the starting point is closer to the minimum

```python
gradient_descent(fx,1.5,0.1,0.000001)
```

```python
gradient_descent(fx,-0.5,0.1,0.000001)
```

Due to the shape(curvature) of the curve(function). when the algorithm approches 1, the steps reduce and smaller steps are taken. But with the larger number i.e xo = 2 larger steps are taken

```python
gradient_descent(fx,-0.5,0.15,0.000001)
```

As we can see when xo = -0.5 and alpha = 0.15, the algorithm gets stuck in a forward backward loop

```python
first = diff_fx(-0.5)
first
```

Implementing the backtracking line search to compute for optimal step size

```python
def backtracking_line_search(fx,xo,diff_fx,alpha,beta):
    mu = 1
    count =0
    while ((fx(xo-mu*diff_fx(xo)))>fx(xo)- alpha*mu*(diff_fx(xo)**2)):
            mu = beta * mu
            count = count +1
    print("Number of iterations:",count,"\nMu(step size) is equal to:")
    return mu                              
```

```python
backtracking_line_search(fx,2,diff_fx,0.1,0.5)
```

```python
backtracking_line_search(fx,1.5,diff_fx,0.1,0.5)
```

```python
backtracking_line_search(fx,-0.5,diff_fx,0.1,0.5)
```

```python
backtracking_line_search(fx,2,diff_fx,0.95,0.5)
```

```python
backtracking_line_search(fx,1.5,diff_fx,0.95,0.5)
```

```python
backtracking_line_search(fx,-0.5,diff_fx,0.95,0.5)
```

#### From the above instances we observe that, the larger the alpha the smaller the step size(mu) and vice-versa. alpha selection is imperative in computation of mu(step size)
